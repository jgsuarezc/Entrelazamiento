Probabilidad conjunta de detectar fotones con polarización HH o VV, dado que la polarización de B es $\theta_B$ y la polarización de A es $\theta_A = 45^\circ$

\begin{equation}
  P(45,\theta_{B})=\frac{1}{4}(1+2sen\theta_{B} cos\theta_{B})
\end{equation}

```python


# Genera datos para el eje x en grados (por ejemplo, de -45 grados a 360 grados)
x_grados = np.linspace(-45, 360, 1000)

# Convierte los grados a radianes
x_radianes = np.deg2rad(x_grados)

# Calcula los valores de la función según la fórmula
p = (1/4) * (1 + 2 * np.sin(x_radianes) * np.sin(x_radianes))

# Crea el gráfico
plt.plot(x_grados, p)

# Personaliza el gráfico
plt.xlabel(r'$\theta_2$ (grados)')
plt.ylabel('probabilidad ')
plt.title(r'Gráfico de p($\theta_1$=45,$\theta_2$) ')
plt.legend()
plt.grid(True)

# Muestra el gráfico
plt.show()
```
